```python
import requests
from bs4 import BeautifulSoup
```

```python

# TAREFA 1 (Scraping de Websites com BeautifulSoup)
# URL da página a ser analisada
url = "https://www.nationalgeographicbrasil.com/"

# Faz a requisição HTTP para obter o conteúdo da página
response = requests.get(url)

# Verifica se a requisição foi bem-sucedida
if response.status_code == 200:
    # Cria o objeto BeautifulSoup para parsear o HTML
    soup = BeautifulSoup(response.text, 'html.parser')

    # Extrai e imprime o título da página
    title = soup.title.string
    print("Título da página:", title)

    # Extrai e imprime todos os links da página
    print("\nLinks encontrados:")
    for link in soup.find_all('a', href=True):
        print(link['href'])
else:
    print("Erro ao acessar a página:", response.status_code)

```

```python
# TAREFA 2 (Extração de Texto de Imagens com OCR)

# Importa a biblioteca pytesseract
import pytesseract
from PIL import Image
```

```python


# Caminho da imagem
image_path = 'dn2.png'

# Carrega a imagem usando PIL
image = Image.open(image_path)

# Extrai o texto da imagem usando pytesseract
texto_extraido = pytesseract.image_to_string(image)

# Caminho para salvar o arquivo de texto
output_path = 'resultado.txt'

# Salva o texto extraído em um arquivo .txt
with open(output_path, 'w') as arquivo:
    arquivo.write(texto_extraido)

print("Texto extraído e salvo em", output_path)

```

```python
# TAREFA 3 (Implementação Completa de um Extrator de Dados Estruturados e Não Estruturados)
import fitz  # PyMuPDF
```

```python

def extract_from_html(url):
    print("\nExtraindo dados de HTML:")
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        title = soup.title.string if soup.title else "Sem título"
        links = [link['href'] for link in soup.find_all('a', href=True)]

        print("Título:", title)
        print("Links:")
        for link in links:
            print("-", link)
    else:
        print("Erro ao acessar a página:", response.status_code)

def extract_from_pdf(pdf_path):
    print("\nExtraindo dados de PDF:")
    with fitz.open(pdf_path) as pdf:
        text = ""
        for page_num in range(len(pdf)):
            text += pdf[page_num].get_text()

        print("Texto extraído do PDF:")
        print(text)

def extract_from_image(image_path):
    print("\nExtraindo dados de Imagem:")
    image = Image.open(image_path)
    text = pytesseract.image_to_string(image)

    print("Texto extraído da imagem:")
    print(text)

def main(file_path):
    # Identifica o tipo de arquivo com base na extensão
    if file_path.startswith("http"):
        extract_from_html(file_path)
    elif file_path.endswith('.pdf'):
        extract_from_pdf(file_path)
    elif file_path.endswith(('.jpg', '.jpeg', '.png')):
        extract_from_image(file_path)
    else:
        print("Tipo de arquivo não suportado")

# Exemplo de URLs e caminhos de arquivos
url = "https://www.nationalgeographicbrasil.com/"
pdf_path = "poemas_classicos_chinese_trecho.pdf"
image_path = "dn2.png"

# Execução para cada tipo de arquivo
main(url)
main(pdf_path)
main(image_path)
```
